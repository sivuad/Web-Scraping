```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://hotwheels.fandom.com/wiki/Fast_%26_Furious_Series_(2023)'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
kata_kunci = table.find_all('th')
```

```python
kata_kunci
```

```python
kata_kunci_table = [kunci.text.strip() for kunci in kata_kunci]

print(kata_kunci_table)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = kata_kunci_table)
df
```

```python
columns_data = table.find_all('tr')
```

```python
for row in columns_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
